```python
import sys
import numpy
numpy.set_printoptions(threshold=sys.maxsize)
#Problem 31 Coin Sums
A = np.array([1,2,5,10,20,50,100,200])
for i in range(1,201):
    A  = np.vstack((A,np.array([1,0,0,0,0,0,0,0])))
    for j in range(1,8):
        if i<A[0,j]:
            A[i,j] = A[i,j-1]
        elif i==A[0,j]:
            A[i,j]=1+A[i,j-1]
        else:
            A[i,j]=A[i,j-1]+A[(i-A[0,j]),j]
print(A)
```
