Pentagonal numbers are generated by the formula, Pn=n(3n−1)/2. The first ten pentagonal numbers are:

1, 5, 12, 22, 35, 51, 70, 92, 117, 145, ...

It can be seen that P4 + P7 = 22 + 70 = 92 = P8. However, their difference, 70 − 22 = 48, is not pentagonal.

Find the pair of pentagonal numbers, Pj and Pk, for which their sum and difference are pentagonal and D = |Pk − Pj| is minimised; what is the value of D?

```python
def pentagonal_generator(n):
    listy = []
    for i in range(1,n + 1):
        listy += [int((i*(3*i - 1))/2)]
    return listy
```

```python
first_ten = pentagonal_generator(10)
```

```python
print(first_ten)
```

```python
pentagonal_sum = []
for element in first_ten:
    for element_2 in first_ten:
        if ((element + element_2) in first_ten) == True:
            pentagonal_sum += [(element, element_2)]
```

```python
def check_list(list_1):
    list_2 = []
    list_3 = []
    for i in range(0,len(list_1)):
        for j in range (i, len(list_1)):
            if (list_1[i]+list_1[j] in list_1) == True and (list_1[j]-list_1[i] in list_1) == True:
                list_2 += [(list_1[i],list_1[j])]
                list_3 += [list_1[j]-list_1[i]]
    return list_2, list_3
```

```python
tester_list = pentagonal_generator(10000)
print(check_list(tester_list))
```

```python
def pofk(m,n):
    pofk = (3*(n**2)-n-3*(m**2)+m)/4
    intpofk = int(pofk)
    return intpofk
```

```python
print(pofk(1,2))
```

```python
def pofj(m,n):
    pofk_= pofk(m,n)
    pofj_ = .5*m*(3*m - 1) + pofk_
    return pofj_
```
