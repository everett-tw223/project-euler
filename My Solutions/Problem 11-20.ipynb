Problem 12 

The sequence of triangle numbers is generated by adding the natural numbers. So the 7th triangle number would be 1 + 2 + 3 + 4 + 5 + 6 + 7 = 28. The first ten terms would be:

1, 3, 6, 10, 15, 21, 28, 36, 45, 55, ...

Let us list the factors of the first seven triangle numbers:

 1: 1
 3: 1,3
 6: 1,2,3,6
10: 1,2,5,10
15: 1,3,5,15
21: 1,3,7,21
28: 1,2,4,7,14,28
We can see that 28 is the first triangle number to have over five divisors.

What is the value of the first triangle number to have over five hundred divisors?

```python
def generatetriangle(input1):
    i=1
    output = 0
    while i<=input1:
        output+=i
        i+=1
    return output
print(generatetriangle(7))
```

```python
import math
def numdivisors(input1):
    output=0
    if int(math.sqrt(input1))**2==input1:
        output+=1
    i=1
    while i<math.sqrt(input1):
        if input1%i==0:
            output+=2
        i+=1
    return output
print(numdivisors(28))
            
    #divisors come in pairs with the singular exception of the square root
```

```python
i = 1
while numdivisors(generatetriangle(i))!=500:
    i+=1
print(generatetriangle(i))
```

```python
i=1
maxnum=0
beta=0
target = 500
while maxnum<target:
    beta+=i
    alpha=numdivisors(beta)
    if alpha>maxnum:
        maxnum=alpha
    i+=1
print(beta)
```

```python
def collatz(x):
    n=1
    while x!=1:
        if x%2==0:
            x=x/2
        elif x%2==1:
            x=(3*x)+1
        n+=1
    return n
print(collatz(13))
```
